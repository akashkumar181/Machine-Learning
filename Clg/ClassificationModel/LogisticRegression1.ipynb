```python
# üçã 1Ô∏è‚É£ Precision ‚Äî Kitno ka test sahi nikla?

# üëâ Imagine: Doctor bola 10 log COVID-positive hain.
# But sach me sirf 8 logon ko hi COVID tha.

# üßæ So precision = 8/10 = 80%

# üó£Ô∏è Easy line:

# ‚ÄúModel ne jitno ko COVID bola, unme se 80% log waqai me positive the.‚Äù

# üí¨ Real life: Precision high matlab model jhoot kam bolta hai.
# ("Galat se positive declare kam karta hai.")
```

```python
# üçã 2Ô∏è‚É£ Recall ‚Äî Kitno ko pakda model ne?

# üëâ 10 logon ko actually COVID tha,
# but model ne sirf 8 logon ko hi pakda.

# üßæ So recall = 8/10 = 80%

# üó£Ô∏è Easy line:

# ‚ÄúJitno ko sach me COVID tha, unme se 80% ko model ne pakad liya.‚Äù

# üí¨ Real life: Recall high matlab model kisi sick bande ko miss nahi karta.
# ("Jo bimaar hai, use pakad lega.")
```

```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Step 2: Sample dataset
data = {
    'age': [22, 25, 28, 32, 36, 40, 45, 48, 52, 57],
    'income': [35000, 40000, 48000, 50000, 55000, 60000, 62000, 70000, 75000, 80000],
    'buys_insurance': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'yes']
}

df = pd.DataFrame(data)

# Step 3: Convert target into numeric (yes=1, no=0)
df['buys_insurance'] = df['buys_insurance'].map({'yes': 1, 'no': 0})

# Step 4: Define features and target
X = df[['age', 'income']]
y = df['buys_insurance']

# Step 5: Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 6: Create and train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Step 7: Predict on test data
y_pred = model.predict(X_test)

# Step 8: Evaluate model
print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))
print("\nüîπ Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nüîπ Classification Report:\n", classification_report(y_test, y_pred))

# Step 9 (Optional): Predict probability
print("\nüîπ Predicted Probabilities:\n", model.predict_proba(X_test))
```

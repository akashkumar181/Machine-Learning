```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
data={
    "Age":[18,19,20,21,18,22,23,20,19,22],
    "StudyHour":[2,3,4,5,1,6,3,5,4,2],
    "Pass":["No","No","Yes","Yes","No","Yes","No","Yes","Yes","No"]}
```

```python
df=pd.DataFrame(data)
df
```

```python
X=df[['Age','StudyHour']]
Y=df['Pass']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=6,random_state=42)
model.fit(X_train,Y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
print(y_predict)
```

```python
acc = accuracy_score(Y_test, y_predict)
print(acc)
```

```python
plot_tree(model,feature_names=['Age','StudyHours'],class_names=['Yes','No'],filled=True)
plt.show()
```

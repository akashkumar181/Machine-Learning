```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
e=[1,5,6,0,4,9]
s=[20000,1000000,120000,15000,80000,150000]
```

```python
#Loading the Dataset
df=pd.DataFrame({
    'exp':e,
    'sal':s
})
df
```

```python
#Split the data into features(x) and target (y)
x=df[['exp']]
y=df['sal']
```

```python
#Linear Regression
model=LinearRegression()
model.fit(x,y)
#predication
```

```python
#predication
y_pred=model.predict(x)
```

```python
#plot regression
plt.figure(figsize=(12,5))
```

```python
#without scaling
plt.scatter(x,y,color='g',label='Actaul')
plt.plot(x,y_pred,color='r',label='Predicated')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#print comparsion
print("Model without scaling")
print(f"Coefficient(Slope):{model.coef_[0]:.2f}")
print(f"Intercept::{model.intercept_:.2f}")
#predict for new value 7.5
new_value=np.array([[7.5]])
predicated_salary=model.predict(new_value)
print("Predicted salary for 7.5 years experience: {predicted_salary[0]:.2f}",predicated_salary)
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
data = {
    'area_sqft': [1000, 1500, 2000, 2500, 3000, 3500, 4000],
    'bedrooms': [2, 3, 3, 4, 4, 5, 5],
    'price': [200000, 250000, 280000, 310000, 360000, 400000, 420000]
}
```

```python

df = pd.DataFrame(data)
```

```python
# Define features (X) and target (y)
X = df[['area_sqft', 'bedrooms']]
y = df['price']
```

```python

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict prices
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("=== Multiple Linear Regression Results ===")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R² Score: {r2:.2f}")

# Show predictions
results = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred})
print("\nPredicted vs Actual:")
print(results)
```
